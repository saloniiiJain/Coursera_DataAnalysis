```python
#import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use('bmh')
```

```python
df = pd.read_csv('/content/Data.csv',header=[0])
df.head(10)
```

```python
df.info()
```

```python
df=df.dropna(axis=1,thresh=3, how='all')
```

```python
df.head()
```

```python
df.info()
```

```python
df["Enrolled"] = pd.to_numeric(df["Enrolled"])
df["Instructors_No"] = pd.to_numeric(df["Instructors_No"])
df["People_rated"] = pd.to_numeric(df["People_rated"])
df["Views"] = pd.to_numeric(df["Views"])
```

```python
plt.figure(figsize=(20,10))
plt.title("Domain")
plt.xticks(rotation = 90)
sns.countplot(df['Domain'])
```

```python
plt.figure(figsize=(10,10))
plt.pie(df['Proffeciency Level'].value_counts(),labels=df['Proffeciency Level'].unique(),startangle = 90,explode=[0,0,0.1],autopct='%.2f%%' )
plt.show() 
```

```python
print(df['Rating'].describe())
plt.figure(figsize=(9, 8))
sns.distplot(df['Rating'], color='g', bins=100, hist_kws={'alpha': 0.4});
```

### **Bivariate Analysis**

```python
# plotting heatmap usingl all methods for all numerical variables (peason, kendall, spearman)
#Correlation matrix
plt.figure(figsize=(36,6), dpi=140)
for j,i in enumerate(['pearson','kendall','spearman']):
  plt.subplot(1,3,j+1)
  correlation = df.dropna().corr(method=i)
  sns.heatmap(correlation, linewidth = 2)
  plt.title(i, fontsize=18)
```

```python
import missingno as msno
msno.heatmap(df)
```

```python
#The dendrogram allows you to more fully correlate variable completion, revealing trends deeper than the pairwise ones visible in the correlation heatmap
msno.dendrogram(df)
```
